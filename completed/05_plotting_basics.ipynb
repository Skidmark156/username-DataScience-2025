## 1. Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

âœ… **Your Turn**: Plot `cos(x)` on the same graph with a different color and label.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", color="blue")
plt.plot(x, y2, label="cos(x)", color="orange")
plt.title("Line Plot with sin(x) and cos(x)")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.show()
```

## 3. Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

âœ… **Your Turn**: Generate two sets of random points and plot them in different colors on the same scatter plot.

```python
# First group
x1 = np.random.rand(50)
y1 = np.random.rand(50)

# Second group
x2 = np.random.rand(50)
y2 = np.random.rand(50)

plt.scatter(x1, y1, color="green", label="Group 1", alpha=0.7)
plt.scatter(x2, y2, color="purple", label="Group 2", alpha=0.7)

plt.title("Two Sets of Random Points")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## 4. Bar Plot

```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

âœ… **Your Turn**: Create your own bar chart with at least 4 categories (e.g., favorite foods, courses, or hobbies).

```python
categories = ["Pizza", "Burgers", "Pasta", "Sushi"]
votes = [12, 8, 15, 10]

plt.bar(categories, votes, color=["red", "blue", "green", "orange"])
plt.title("Favorite Foods")
plt.xlabel("Food")
plt.ylabel("Votes")
plt.show()
```

## 5. Customizing Plots

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

âœ… **Your Turn**: Customize your own plot with labels, legend, and grid lines.

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(2*x)   # doubled frequency
y2 = np.cos(3*x)   # tripled frequency

plt.plot(x, y1, label="sin(2x)", linestyle="--", color="blue", linewidth=2)
plt.plot(x, y2, label="cos(3x)", linestyle="-.", color="red", linewidth=2)

plt.title("Customized Trigonometric Plot")
plt.xlabel("x (radians)")
plt.ylabel("Function Value")
plt.legend()
plt.grid(True, linestyle=":", alpha=0.7)
plt.show()
```
