```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
#penguins = sns.load_dataset("penguins")
penguins = pd.read_csv("../../data/penguins.csv")
penguins.head()
```

## 1. Explore Distributions

First, letâ€™s look at the distribution of **body mass**.

```python
sns.histplot(data=penguins, x="body_mass_g", kde=True, hue="species")
plt.title("Distribution of Body Mass by Species")
plt.show()
```

ðŸ‘‰ **Task:** Do some species have higher average body mass than others?

Yes Gentoo penguins tend to have a noticeably higher average body mass compared to Adelie and Chinstrap penguins. The histogram shows the Gentoo distribution shifted to the right.

## 2. Relationships Between Variables

Scatterplots help us see trends between two quantitative variables.

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Flipper Length vs Body Mass")
plt.show()
```

ðŸ‘‰ **Task:** Which species tends to have longer flippers AND higher body mass?

Gentoo penguins again show the trend. They cluster in the upper-right corner, indicating both longer flippers and heavier body mass. Adelie penguins tend to be smaller with shorter flippers.

## 3. Box Plots for Outliers

Box plots are great for spotting outliers.

```python
sns.boxplot(data=penguins, x="species", y="bill_length_mm", palette="Set2")
plt.title("Bill Length by Species")
plt.show()
```

ðŸ‘‰ **Task:** Can you spot any outliers in the bill length data?

Yes there are a few points outside the whiskers, which represent outliers. Penguins with unusually long or short bills for their species.

## 4. Correlation Heatmap

Correlation heatmaps let us see overall variable relationships.

```python
corr = penguins.dropna().corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

Flipper length and body mass show a strong positive correlation longer flippers are generally linked to heavier penguins. Bill length and bill depth have a moderate correlation, depending on species.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset("tips")

# Trend 1: Relationship between total bill and tip
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Total Bill vs Tip Amount")
plt.show()

# Trend 2: Average tip by day
sns.barplot(data=tips, x="day", y="tip", ci="sd", palette="muted")
plt.title("Average Tip by Day")
plt.show()

# Outliers: Tip percentages
tips["tip_pct"] = tips["tip"] / tips["total_bill"] * 100
sns.boxplot(data=tips, x="day", y="tip_pct", palette="Set3")
plt.title("Tip Percentage by Day")
plt.show()
```
