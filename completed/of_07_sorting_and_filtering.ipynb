# ðŸ§® Sorting & Filtering in Pandas

## ðŸ”¹ LEARNING GOALS:
- Sort rows using `.sort_values()`
- Filter data using conditional logic
- Chain multiple filters together
- Use `.query()` for readable filtering


### ðŸ“¥ 1. Load the Dataset

```python
import pandas as pd

df = pd.read_csv("../../data/students.csv")
df.head()
```

### ðŸ”¢ 2. Sorting Data

```python
# Sort by math_score
df.sort_values(by="math_score")
```

```python
# Sort by science_score descending
df.sort_values(by="science_score", ascending=False)
```

```python
# Sort by multiple columns
df.sort_values(by=["science_score", "math_score"], ascending=[False, True])
```

### ðŸ§ª 3. Filtering with Conditions

```python
# Who scored at least 90 in math?
df[df["math_score"] >= 90]
```

```python
# Who scored less than 70 in either subject?
df[(df["math_score"] < 70) | (df["science_score"] < 70)]
```

### ðŸ”— 4. Combining Filters

```python
# Students with B+ or higher in both
df[(df["math_score"] >= 87) & (df["science_score"] >= 87)]
```

### ðŸ§¾ 5. Using `.query()` for Readability

```python
# Same query using .query()
df.query("math_score >= 87 and science_score >= 87")
```

### ðŸ§ª Try It Yourself

- Sort the dataset by first name alphabetically
- Find students with the same math and science score
- Use `.query()` to find students who scored over 95 in *either* subject


```python
# Upload students.csv and run all tasks
from google.colab import files
import pandas as pd

print(" Please upload your students.csv file...")
uploaded = files.upload()

# Load dataset
df = pd.read_csv(list(uploaded.keys())[0])

# Clean column names
df.columns = [c.replace(" ", "_") for c in df.columns]

print(" File loaded!")

# Sort the dataset by first name alphabetically
sorted_by_first = df.sort_values(by="first_name")

# Find students with the same math and science score
same_scores = df[df["math_score"] == df["science_score"]]

# Use .query() to find students who scored over 95 in either subject
high_scores = df.query("math_score > 95 or science_score > 95")

# Show results
print("\n Sorted by first name:")
print(sorted_by_first.head())

print("\n Students with SAME math and science score:")
print(same_scores)

print("\n Students scoring >95 in math OR science:")
print(high_scores)
```

### ðŸ§  Mini-Challenge

> Load `"data/survey.csv"` and:
- Sort by the most recent signup date
- Filter out anyone who didn't complete the survey (nulls or "N/A")
- Use `.query()` to select rows with a specific condition of your choice


```python
# Upload survey.csv and run the full Mini-Challenge
from google.colab import files
import pandas as pd

print(" Please upload your survey.csv file...")
uploaded = files.upload()

# Load dataset
df = pd.read_csv(list(uploaded.keys())[0])

# Clean column names (remove spaces just in case)
df.columns = [c.replace(" ", "_") for c in df.columns]

print(" File loaded!")
print(df.head())

# Sort by most recent signup date
# (Convert date column to datetime format)
df["signup_date"] = pd.to_datetime(df["signup_date"], errors="coerce")
df = df.sort_values(by="signup_date", ascending=False)

# Filter out people who did NOT complete the survey (null or "N/A")
if "completed" in df.columns:
    df = df.dropna(subset=["completed"])  # remove NaN
    df = df[df["completed"].astype(str).str.upper() != "N/A"]  # remove "N/A"

# .query() with a condition of your choice
# Example: find people aged 25 or older
query_result = df.query("age >= 25")

# Show results
print("\n Sorted by most recent signup date:")
print(df.head())

print("\n After removing non-completed surveys:")
print(df.head())

print("\n Query result (age >= 25):")
print(query_result.head())
```
